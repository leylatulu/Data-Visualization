```python
x = np.linspace(0,10, 100)
```

## LINE PLOT

```python
plt.plot(x, np.sin(x),'-')
plt.plot(x, np.cos(x), '--')
plt.xlabel('x ekseni')
plt.ylabel('y ekseni')
plt.legend(('sin', 'cos'))
```

```python
fig = plt.figure()
plt.subplot(2, 1, 1)
plt.plot(x, np.sin(x),'-')
plt.xlabel('x ekseni')
plt.ylabel('y ekseni')
plt.legend(('sin'))

plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x), '--', linewidth=5, color='purple')
plt.xlabel('x ekseni')
plt.ylabel('y ekseni')
plt.legend(('cos'))
```

```python
import seaborn as sns 
import pandas as pd
```

```python
iris = sns.load_dataset('iris')
iris
```

```python
iris['petal_length'].plot()
```

```python
sns.lineplot(data=iris, y='sepal_length', x=iris.index, hue= 'species') 
```

## SCATTER PLOT

```python
fig, ax = plt.subplots(figsize=(5,5)) # dimension creation

ax.scatter(iris['sepal_length'], iris['sepal_width'])
ax.set_title('Iris Dataset')
ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_width')
```

## BAR PLOT

```python
iris.groupby(by='species').mean() #max() ,min(), sum() ...
```

```python
iris.groupby(by='species').mean()['sepal_length'] 
```

```python
data  = iris.groupby(by='species').mean()['sepal_length'] 
ax = sns.barplot(x = data.index, y = data.values)
ax.set_title('Sepal Length Means')
ax.set_ylabel('Max')
```

## HISTOGRAM

```python
sns.histplot(data=iris, y='sepal_width')
```

```python
sns.histplot(data=iris, x='sepal_width', hue='species', bins=20) # bins : allows us to specify ranges
```

```python
sns.histplot(data=iris, x='sepal_width', hue='species', bins=10)
```

```python
iris.corr()
```

## HEATMAP

```python
sns.heatmap(iris.corr(),annot=True) 
# annot : show correlation values
```
